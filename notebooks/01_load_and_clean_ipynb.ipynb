```python
# ==========================================================
# 01 - Limpieza y enriquecimiento del dataset
# ==========================================================

import pandas as pd
import re

# 1. Load original dataset
df = pd.read_csv("../data/raw/reviews.csv")

# ==========================================================
# 2. Basic standardization
# ==========================================================

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"[^a-zA-Z0-9\s.,?!']", " ", text)  # Clean up unusual characters
    text = re.sub(r"\s+", " ", text).strip()
    return text

df["review"] = df["review"].apply(clean_text)
df["label"] = df["label"].str.strip()

# keep the label column
if "label" not in df.columns:
    raise ValueError("ERROR: The file reviews.csv does not have a 'label' column..")

# ==========================================================
# 3. Save clean and expanded dataset
# ==========================================================

df.to_csv("../data/processed/clean_reviews_v2.csv", index=False)

print("\nClean dataset saved as clean_reviews_v2.csv")
print("Final size:", len(df))
```
